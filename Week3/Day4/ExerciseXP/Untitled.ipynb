<h1 style="text-align:left; color:green;">🌟 Exercise 2: Creating a Line Plot for Temperature Variation
</h1>

```python
import matplotlib.pyplot as plt

# Temperature values for each day of the week
temperatures = [72, 74, 76, 80, 82, 78, 75]

# Days of the week
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

# Create the line plot
plt.plot(days, temperatures, marker='o', color='orange', linestyle='-')

# Add labels and title
plt.xlabel('Day')
plt.ylabel('Temperature (°F)')
plt.title('Temperature Variation Over a Week')

# Display the plot
plt.show()
```

<h1 style="text-align:left; color:green;">🌟 Exercise 3: Visualizing Monthly Sales with a Bar Chart
</h1>

```python
import matplotlib.pyplot as plt

# Sales values for each month
sales = [5000, 5500, 6200, 7000, 7500]

# Corresponding month labels
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

# Create the bar chart
plt.bar(months, sales, color='skyblue')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales for Retail Store')

# Display the plot
plt.show()
```

<h1 style="text-align:left; color:green;">🌟 Exercise 4: Visualizing the Distribution of CGPA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

 
df = pd.read_csv('Student Mental health.csv')

 
sns.histplot(df['What is your CGPA?'], bins=5, color='teal', kde=False)

 
plt.title("Distribution des CGPA des étudiants")
plt.xlabel("CGPA")
plt.ylabel("Nombre d'étudiants")

 
plt.show()
```

<h1 style="text-align:left; color:green;">🌟  Exercise 5: Comparing Anxiety Levels Across Different Genders
</h1>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

 
sns.countplot(
    data=df,
    x='Choose your gender',
    hue='Do you have Anxiety?',
    palette='Set2'
)

 
plt.title("Comparaison des niveaux d'anxiété selon le genre")
plt.xlabel("Genre")
plt.ylabel("Nombre d'étudiants")

 
plt.show()
```

<h1 style="text-align:left; color:green;">🌟Exercise 6: Exploring the Relationship Between Age and Panic Attacks
</h1>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 

# Convertir les réponses 'Do you have Panic attack?' en valeurs numériques : Yes=1, No=0
df['Panic_numeric'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

# Créer le scatter plot
sns.scatterplot(
    data=df,
    x='Age',
    y='Panic_numeric',
    hue='Panic_numeric',
    palette={0:'blue', 1:'red'},
    alpha=0.6,
    s=100
)

# Ajouter labels et titre
plt.title("Relation entre l'âge et les attaques de panique")
plt.xlabel("Âge")
plt.ylabel("Panic Attack (0=Non, 1=Oui)")

# Afficher le graphique
plt.show()
```
