<h1 style="text-align:left; color:green;">Exercise 2: Transformation Exercise
</h1>

Purpose : suggest a method to transform into a structured data

### A series of blog posts about travel experiences.- method : NLP (Natural Language Processing)
- Why? Blog posts are texts (unstructered data), with NLP we can transform raw text into tables or JSON with meaningful attributes.
- How? Use text cleaning, tokenization, and named entity recognition(NER) with Scipy, in consequence, wa can extract structred fields like destination, dates, activities, sentiment and costs.ng.

### Audio recordings of customer service calls.
- method : Speech to text  + NLP
- Why? audios can't be analyzed as structured data, Transcription enables further structuring
- analysis.
- How? convert audio to text using speech recognition and apply NLP to identify issues, keywords, call duration, customer sentiment

### Handwritten notes from a brainstorming session.
- method: OCR (Optical character Recognition) + text processing
- Why? handwritten content must be digitized first. OCR transform it into text that can be then structured
- How? use OCR to digitize notes and organize content into ideas, themes, action items.

### A video tutorial on cooking.
- method: Computer vision + speech to text
- Why? videos combine visual and audio data. we use computer vision to analyse frames and use the transcription to anlyse tructered data.
- How? Extract audio and transcribe spoken instructions, also use computer visio to detect ingredients, steps, timestamps.
