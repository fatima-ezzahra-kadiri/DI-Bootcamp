### What is data analysis?

Data analysis is the process of cleaning, transforming, and modeling data to uncover patterns, explain trends, and support better decision-making.

### Why is data analysis important in modern contexts?

Because today we generate huge amounts of data, and analyzing it helps people and organizations make smarter, faster, and more accurate decisions.

### three areas where data analysis is applied today

- Healthcare : Used to predict diseases, improve treatments, and optimize hospital operations.

- Business and Marketing : Helps understand customer behavior, improve products, and increase sales.

- Finance : Used to detect fraud, manage risk, and support investment decisions.

<h1 style="text-align:left; color:green;">Exercise 3: Identifying Data Types
</h1>

```python
health_df["Schizophrenia (%)"].unique()
```

```python
import re

unique_vals = health_df["Schizophrenia (%)"].unique()

pattern = re.compile(r"[NnAa ]")  # cherche N, n, A, a ou espace
suspicious_vals = [val for val in unique_vals if isinstance(val, str) and pattern.search(val)]

print(suspicious_vals)
```

```python
import re

unique_vals = health_df["Bipolar disorder (%)"].unique()

pattern = re.compile(r"[NnAa ]")  # cherche N, n, A, a ou espace
suspicious_vals = [val for val in unique_vals if isinstance(val, str) and pattern.search(val)]

print(suspicious_vals)
```

certaines colonnes censées contenir des valeurs numériques sont détectées par Pandas comme type objet, Cela arrive parce qu’elles contiennent aussi des valeurs textuelles descriptives.

<h1 style="text-align:left; color:green;">Exercise 5: Basic Data Analysis with Google Colab</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\pc\Downloads\Iris Dataset\Iris.csv")
data.head()
```

```python
column = data['SepalLengthCm']

mean_value = column.mean()
median_value = column.median()
mode_value = column.mode()[0]

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
len(column.unique())
```

```python
plt.hist(column, bins=10)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
sns.histplot(column, bins=10, kde=True, stat="count")
#sns.histplot(column, bins=10, kde=False)
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(data['SepalWidthCm'], data['Species'])
plt.show()
```
