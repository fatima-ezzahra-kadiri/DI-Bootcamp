# Exercise 1: Comparative Analysis of Generative AI and Traditional AI

## 1. Automated Medical Diagnosis
A **traditional AI classification model** is more suitable for lung cancer detection from X-ray images. Classification models such as CNNs are highly optimized for image recognition tasks and can provide high accuracy when trained on labeled medical datasets. Generative AI could be used as a complementary tool to **augment the dataset by generating synthetic medical images**, which may help improve model performance when real data is limited. However, generative models may introduce unrealistic patterns that could reduce reliability in medical diagnosis. Since healthcare systems require high accuracy and explainability, traditional AI models are generally preferred.

## 2. Legal Document Generation
A **generative AI language model** would be more suitable because drafting contracts requires generating structured natural language. Generative models can analyze large corpora of legal documents and produce drafts quickly, improving efficiency. However, there are risks such as **incorrect legal interpretations, hallucinated clauses, and potential liability issues**. For this reason, generative AI should be used as a **drafting assistant**, while final validation must be performed by legal professionals.

## 3. AI-Generated Scientific Research
Generative AI can be useful for **summarizing large numbers of research papers and identifying patterns across literature**. It can significantly reduce the time needed for literature reviews and help researchers discover connections between topics. However, generative models may sometimes produce **hallucinated citations or inaccurate summaries**. Therefore, generative AI should support researchers but not replace human validation and critical analysis.

## 4. Financial Market Predictions
For financial forecasting, **traditional AI models such as regression, ARIMA, or time-series forecasting** are generally more appropriate. These models are specifically designed for numerical prediction and can capture trends and seasonality in historical data. Generative AI could potentially simulate market scenarios, but it is less reliable for precise forecasting. Financial decisions require **high reliability and interpretability**, which traditional statistical models often provide better.

## 5. Autonomous Vehicle Decision-Making
Traditional AI models are preferred for **real-time decision-making in autonomous vehicles**. Systems based on computer vision, reinforcement learning, and sensor fusion are optimized for safety-critical environments. Generative AI could be used to **simulate driving environments for training purposes**, but it should not directly control vehicle decisions because unpredictability could lead to safety risks.

# Exercise 2: Ethical and Security Risks of Generative AI

**Risks**
- Spread of misinformation during elections
- Manipulation of public opinion
- Loss of trust in digital media

**Consequences**
If left unregulated, deepfakes could destabilize democratic processes and increase political polarization.

**Solutions**
- AI-based deepfake detection systems
- Regulations requiring labeling of AI-generated media

**Risks**

- Multi-factor authentication combining biometrics and behavioral signals

## 2. Bias and Fairness in AI Training Data
To reduce bias, datasets should include **diverse and balanced sources of information** representing different political perspectives and geographic regions. Data should be carefully filtered to remove misleading or biased content. Additionally, regular **bias audits and fairness evaluations** should be conducted to detect and correct skewed outputs.

## 3. Fine-Tuning for Domain-Specific Tasks

**Key steps**

1. **Data collection and preprocessing**
   - Collect high-quality medical research papers
   - Clean and structure the dataset
   - Remove unreliable or outdated sources

2. **Transfer learning**
   - Start with a pretrained language model
   - Fine-tune it using domain-specific medical literature

3. **Evaluation metrics**
   - Factual accuracy
   - Perplexity
   - Human expert review

---

## 4. Evaluating Generative AI Performance

Common quantitative metrics include:

- **BLEU** – evaluates similarity between generated and reference text (often used in translation)
- **Perplexity** – measures how well a language model predicts text
- **FID (Fréchet Inception Distance)** – measures the quality of generated images compared to real images
- **ROUGE** – evaluates summarization quality

---

## 5. Controlling AI Creativity and Coherence

- **Temperature scaling** controls randomness in generation. Lower values produce safer outputs, while higher values increase creativity.
- **Reinforcement Learning from Human Feedback (RLHF)** helps align outputs with human expectations and reduce harmful responses.
- **Attention mechanisms** allow models to focus on relevant parts of the input, improving coherence and contextual understanding.

# Exercise 4: Evaluating the Trade-offs Between GANs and VAEs

## 1. Synthetic Medical Image Generation
**VAEs** are preferable because they provide stable training and structured latent spaces. This makes them useful for generating controlled medical images while preserving privacy.

## 2. AI-Assisted Creative Writing
**VAEs** can help learn structured latent representations of text styles, enabling smoother interpolation between writing styles.

## 3. Anomaly Detection in Financial Transactions
**VAEs** are commonly used because they learn the distribution of normal data and detect anomalies when reconstruction error is high.

## 4. Generating High-Resolution Fashion Designs
**GANs** are more suitable because they produce highly realistic and detailed images, making them ideal for fashion design generation.

## 5. Data Augmentation for Autonomous Vehicles
**GANs** are preferred since they can generate realistic driving scenarios and environmental variations for training perception models.

# Exercise 5: Advanced Latent Space Exploration in VAEs

## 1. Visualizing Latent Space Distributions
To evaluate whether digits are separated in latent space, clustering techniques such as **k-means or t-SNE visualization** can be applied. These methods help visualize how different digit classes are distributed in the latent space.

---

## 2. Interpolating Between Two Samples
To interpolate between digits like "3" and "8", encode both images into latent vectors and linearly interpolate between them. Decoding intermediate vectors generates gradual transformations between digits. VAEs allow smooth interpolation because their latent space is continuous and regularized.

---

## 3. Controlling Variability in Generated Outputs
Increasing the **KL divergence weight** encourages stronger regularization and smoother latent distributions, which increases diversity but may reduce reconstruction quality. Decreasing it produces more accurate reconstructions but less diverse samples.

Example application: **drug molecule generation**, where controlled diversity is needed to explore new chemical structures.

---

## 4. Disentangling Latent Representations
If each latent dimension corresponds to a specific feature, modifying a single dimension could change attributes such as **hair color or facial expression** while keeping other features unchanged. This enables controlled generation of variations.

---

## 5. Comparing PCA and VAEs

**Linearity vs Non-linearity**
- PCA is a linear dimensionality reduction method
- VAEs capture complex nonlinear relationships

**Interpretability**
- PCA components are easier to interpret
- VAE latent variables are more abstract

**Data Reconstruction Quality**
- PCA performs well for simple datasets
- VAEs reconstruct complex data more effectively

**Generative Modeling**
- PCA cannot generate new samples easily
- VAEs are designed specifically for generative modeling
