What you will create
A command-line Tic Tac Toe game that allows two players to take turns marking a 3x3 grid.

Instructions:
Tic Tac Toe is played on a 3x3 grid. Players take turns marking empty squares with their symbol (‘O’ or ‘X’). The first player to get three of their symbols in a row (horizontally, vertically, or diagonally) wins. If all squares are filled and no player has three in a row, the game is a tie.

```python
Game_Board = [[" "," "," "],[" "," "," "],[" "," "," "]]
```

```python
def display_board(Game_Board):
    for vect in Game_Board:
        print(vect)
        
display_board(Game_Board)
```

```python
def player_input(player, Game_Board):
    while True:

       
        while True:
            position_x = input("Enter a row number (1–3): ")
            try:
                position_x = int(position_x)
                if 1 <= position_x <= 3:
                     break
                else:
                    print("Invalid row!")
                 
            except:
                print("enter a number")
             
             
            
 

 
        while True:
            position_y = input("Enter a column number (1–3): ")
            try:
               position_y = int(position_y) 
               if 1 <= position_y <= 3:
                   break
               else:
                   print("Invalid column!")
            except:
                print("enter a number")
                 

 
        x = position_x - 1
        y = position_y - 1

       
        if Game_Board[x][y] == " ":
         
            return x, y
        else:
            print("Cell is not empty, choose another one.")
```

```python
def check_win(board, player):
    for i in range(3):
        if (board[i][0]==board[i][1]==board[i][2]==player) or (board[0][i]==board[1][i]==board[2][i]==player) or (board[1][1]==board[2][2]==board[0][0]==player) or (board[0][2]==board[1][1]==board[2][0]==player):
            #print(f"the player{player}is the winner")
            return True
    return False
```
