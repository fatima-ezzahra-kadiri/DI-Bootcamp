Instructions
You are given two lists. Convert them into a dictionary where the first list contains the keys and the second list contains the corresponding values.



```python
keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]
```

Key Python Topics:

Instructions
Write a program that calculates the total cost of movie tickets for a family based on their ages.

Family members’ ages are stored in a dictionary.
The ticket pricing rules are as follows:
Under 3 years old: Free
3 to 12 years old: $10
Over 12 years old: $15


Family Data:

family = {"rick": 43, 'beth': 13, 'morty': 5, 'summer': 8}


Loop through the family dictionary to calculate the total cost.
Print the ticket price for each family member.
Print the total cost at the end.


Bonus:

Allow the user to input family members’ names and ages, then calculate the total ticket cost.



```python
family = {"rick": 43, 'beth': 13, 'morty': 5, 'summer': 8}
```

```python
total_cost = 0
for key, value in family.items():
    if value < 3:
        cost_member = 0
    elif value > 12:
        cost_member = 15
    else:
        cost_member = 10
    total_cost += cost_member
    print(f"ticket price for {key} is {cost_member}")

print(f"the total cost is {total_cost}")
```

Exercise 4: Disney Characters
Key Python Topics:

Looping with indexes
Dictionary creation
Sorting


Instructions
You are given a list of Disney characters. Create three dictionaries based on different patterns as shown below:

```python
users = ["Mickey", "Minnie", "Donald", "Ariel", "Pluto"]
```

```python
indices = [0, 1, 2, 3, 4]
dict1 = dict(zip(users, indices))
dict1
```

```python
dict2 = dict(zip(indices, users))
dict2
```

```python
sorted_users = sorted(users)
dict3 = {user: i for i, user in enumerate(sorted_users)}
print(dict3)
```

```python
sorted_users = sorted(users)
dict3 = dict(zip(sorted_users, indices))
print(dict3)
```
