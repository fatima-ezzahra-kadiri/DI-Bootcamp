Goal: Create a dictionary that stores the indices (number of the position) of each letter in a word provided by the user(input()).



```python
word = input("enter a word: ")
```

```python
dic_word = {}
for i in range(len(word)):
    letter = word[i]
    if letter in dic_word:
        dic_word[letter].append(i)
    else:
        dic_word[letter]=[i]
dic_word
    
   
```

<h2 style="text-align:left; color:green;">Challenge 2: Affordable Items</h2>

Goal: Create a program that prints a list of items that can be purchased with a given amount of money

```python
#1. Store Data:
items_purchase = {"Apple": "$4", "Honey": "$3", "Fan": "$14", "Bananas": "$4", "Pan": "$100", "Spoon": "$2"}
wallet = "$100" 
```

```python
# items_purchase = {"Phone": "$999", "Speakers": "$300", "Laptop": "$5,000", "PC": "$1200"}
# wallet = "$1"
```

```python
#2. Data Cleaning:
wallet_clean = int(wallet.replace('$', "").replace(',', ""))
wallet_clean
```

```python
items_purchase_clean = {}
for item, price in items_purchase.items():
    price = int(price.replace('$', "").replace(',', ""))
    items_purchase_clean[item] = price
 
items_purchase_clean
```

```python
#3. Determining Affordable Items:
```

```python
basket = []
for item, price in items_purchase_clean.items():
    if price <= wallet_clean:
        basket.append(item)
        wallet_clean -= price
        if wallet_clean == 0:
            break
if len(basket) == 0:
    print("Nothing")
else:
    print(sorted(basket))  
```
