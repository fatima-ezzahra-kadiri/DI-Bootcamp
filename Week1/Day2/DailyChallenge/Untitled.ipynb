```python
number = int(input("enter a number:"))
lenght = int(input("enter a lenght:"))
```

```python
output = [number * i for i in range(1,lenght+1)]
output
```

```python
text = input("enter a string: ")
output = text[0]
for i in range(1, len(text)):
    if text[i] != text[i-1]:
        output += text[i]

print(f"the text without consecutive letters is: {output} ")            
```
