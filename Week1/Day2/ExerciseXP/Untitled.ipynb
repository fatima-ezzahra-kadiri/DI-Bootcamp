```python
liste = []
topping = input("enter pizza toppings one by one: ")

while topping != 'quit':
    liste.append(topping)
    print(f"Adding {topping} to your pizza.")
    topping = input("enter pizza toppings one by one: ")

if len(liste)>0 :
    print(f"The topping are {', '.join(liste)}, and the total cost of the pizza is {len(liste)*2.50 + 10} $")
```

```python
person_number = int(input("enter a number of person"))
Tcost = 0
for i in range(person_number):
    age = int(input(f"enter a age of the {i+1} person"))
    if age < 3:
        Tcost += 0
    elif age > 12:
        Tcost += 15
    else:
        Tcost += 10
print(f"The total cost is {Tcost}")   
```
