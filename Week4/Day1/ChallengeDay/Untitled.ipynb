<h1 style="color:green;">Social Support and Happiness</h1>

```python
#Employ a Matplotlib scatter plot to investigate the relationship between ‘Social support’ and ‘Score’.
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df_2019['Social support'], df_2019['Score'])

plt.title('Relationship Between Social Support and Happiness Score (2019)')
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')

plt.grid(True)
plt.show()
```

<h1 style="color:green;">Regional Comparisons of GDP and Health</h1>

```python
#Create a Matplotlib subplot that compares ‘GDP per Capita’ and ‘Healthy Life Expectancy’ across different regions.
```

```python
regions = df_2019["Country or region"].unique()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14,12))

# 1) GDP per Capita by Region
axes[0].boxplot([df_2019[df_2019["Country or region"] == r]["GDP per capita"] for r in regions])
axes[0].set_xticklabels(regions, rotation=90, fontsize=6)
axes[0].set_title("GDP per Capita across Regions")
axes[0].set_ylabel("GDP per Capita")

# 2) Healthy Life Expectancy by Region
axes[1].boxplot([df_2019[df_2019["Country or region"] == r]["Healthy life expectancy"] for r in regions])
axes[1].set_xticklabels(regions, rotation=90, fontsize=6)
axes[1].set_title("Healthy Life Expectancy across Regions")
axes[1].set_ylabel("Healthy Life Expectancy")

plt.tight_layout()
plt.show()
```

```python
#Use a bar plot for ‘GDP per Capita’ and a line plot for ‘Healthy Life Expectancy’ on the same axes to observe how economic strength relates to health outcomes in different regions.
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(14,6))

# Bar plot pour GDP per Capita
ax1.bar(df_2019["Country or region"], df_2019["GDP per capita"], color='skyblue', label='GDP per Capita')
ax1.set_xlabel("Country or Region")
ax1.set_ylabel("GDP per Capita", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# X_labels verticales et petites
ax1.set_xticks(range(len(df_2019["Country or region"])))  # position des ticks
ax1.set_xticklabels(df_2019["Country or region"], rotation=90, fontsize=6)

# Line plot pour Healthy Life Expectancy
ax2 = ax1.twinx()
ax2.plot(df_2019["Country or region"], df_2019["Healthy life expectancy"], color='red', marker='o', label='Healthy Life Expectancy')
ax2.set_ylabel("Healthy Life Expectancy", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Titre et mise en page
plt.title("GDP per Capita vs Healthy Life Expectancy by Country/Region")
plt.tight_layout()
plt.show()
```
