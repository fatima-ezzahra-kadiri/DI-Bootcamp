 Exercise 2 : Creating a Line Plot for Temperature Variation
Instructions
Create a simple line plot using Matplotlib that represents the temperature variations over a week.
Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]) and label the x-axis as “Day” and the y-axis as “Temperature (°F)”.

```python
import matplotlib.pyplot as plt
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
Temp = [72, 74, 76, 80, 82, 78, 75]

plt.plot(days, Temp, color='blue')
plt.title("Temperature Variation over days of a week")
plt.xlabel("Day")
plt.ylabel("Temperature(°F)")
plt.grid(True)
plt.show()
```

Exercise 3 : Visualizing Monthly Sales with a Bar Chart
Instructions
Generate a bar chart using Matplotlib to visualize the monthly sales data for a retail store.
Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500]) and label the x-axis as “Month” and the y-axis as “Sales Amount ($)”.

```python
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
amount = [5000, 5500, 6200, 7000, 7500]

plt.figure(figsize=(8,5))
plt.bar(months, amount, color='purple')
plt.title("Monthly sales data for a retail store")
plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")
plt.show()
```

Exercise 4 : Data Visualisation
Instructions
Using the dataset: sales_data.csv

Calculate the total quantity of products sold by the company.
Identify the category that had the highest revenue and calculate how much revenue it generated.
Determine the average revenue per sale in the dataset.
Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df["quantity"].dtype
```

```python
#Calculate the total quantity of products sold by the company
total_quantity = df['quantity'].sum()
print(total_quantity)
```

```python
#Identify the category that had the highest revenue and calculate how much revenue it generated.
if all(col in df.columns for col in ['category', 'revenue']):
    revenue_par_category = df.groupby('category')['revenue'].sum()
    highest_revenue_category = revenue_par_category.idxmax()
    highest_revenue = revenue_par_category.max()
    print(f"the category that had the highest revenue is: ${highest_revenue_category} with ${highest_revenue}")
```

```python
#Determine the average revenue per sale in the dataset.
if 'revenue' in df.columns:
    average_revenue_per_sale = df['revenue'].mean()
    print(f"the average revenue per sale is: ${average_revenue_per_sale:.2f}")
```

```python
print(df['date'].dtype)
```

```python
#  Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
    df['quarter'] = df['date'].dt.to_period('Q')
    revenue_each_quarter = df.groupby('quarter')['revenue'].sum()
    plt.figure(figsize=(8,5))
    plt.bar(revenue_each_quarter.index.astype(str), revenue_each_quarter.values)
    plt.title("total revenue generated in each quarter")
    plt.xlabel("Quarter")
    plt.ylabel("total revenue")
    plt.show()
    
```

Exercise 5 : Data Visualisation Using MatPlotLib
Instructions
Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.
Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.

```python
#Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.
import numpy as np
x = np.arange(-10, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.figure()
plt.plot(x,y)
plt.title("the function y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
plt.show()
```

```python
#Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
products = ['A', 'B', 'C', 'D']
sales_values = [15, 30, 45, 20]
plt.figure(figsize=(8,5))
plt.bar(products, sales_values, color='green')
plt.title("Weekly product sales")
plt.xlabel("Product")
plt.ylabel("Sales Values")
plt.show()
```

```python
 #Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
pourcentages = [40, 30, 20, 10]
plt.figure()
plt.pie(pourcentages, labels=fruits, colors=['red', 'orange', 'yellow', 'green'])
plt.title("Favorite fruits distribution")
plt.legend("fruits")
plt.tight_layout()
plt.show()
```
