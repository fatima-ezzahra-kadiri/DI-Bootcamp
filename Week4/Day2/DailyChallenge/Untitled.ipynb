<h3 style="text-align:left; color:'brown';">Data Preprocessing:</h3>

```python
#Handle missing values and convert data types as needed.
```

```python
df["Released_Year"] = pd.to_datetime(
    df["Released_Year"], format="%Y", errors="coerce"
).dt.year.astype("Int64")
```

```python
df["Runtime"] = df["Runtime"].str.replace(" min", "", regex=False).astype("Int64")
```

```python
df["Gross"] = df["Gross"].str.replace(",", "", regex=False)
```

```python
df["Gross"] = pd.to_numeric(df["Gross"], errors="coerce").astype("Int64")
```

```python
df["Gross"] = df["Gross"].fillna(df["Gross"].median())
```

```python
df.head()
```

```python
df.info()
```

```python
df[df["Released_Year"].isna()==True]
```

```python
df.groupby("Certificate").count()
```

```python
most_frequent = df["Certificate"].mode()[0]
most_frequent
```

```python
df["Certificate"] = df["Certificate"].fillna(most_frequent)
```

```python
df = df[df["Released_Year"].notna()]
```

```python
df.loc[:, "Meta_score"] = df["Meta_score"].fillna(df["Meta_score"].mean())
```

<h3 style="text-align:left; color:'brown';">Rating Trends Over Years</h3>

```python
#Create a line plot with Matplotlib showing trends in IMDB_Rating over the years.
```

```python
import matplotlib.pyplot as plt

# Ensure Released_Year is sorted
df_sorted = df.sort_values("Released_Year")

plt.figure(figsize=(12,6))
plt.plot(df_sorted["Released_Year"], df_sorted["IMDB_Rating"])
plt.xlabel("Year")
plt.ylabel("IMDB Rating")
plt.title("IMDB Rating Trend Over the Years")
plt.grid(True)
plt.show()
```

<h3 style="text-align:left; color:'brown';">Star Power Analysis:</h3>

```python
#Employ a Seaborn pairplot to explore relationships involving Stars and their impact on IMDB_Rating and Gross.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

 
top_stars = pd.concat([
    df["Star1"], df["Star2"], df["Star3"], df["Star4"]
]).value_counts().nlargest(5).index

 
for star in top_stars:
    df.loc[:, f"{star}_appears"] = df[["Star1","Star2","Star3","Star4"]].apply(lambda x: int(star in x.values), axis=1)

 
cols_to_plot = [f"{star}_appears" for star in top_stars] + ["IMDB_Rating", "Gross"]

 
sns.pairplot(df[cols_to_plot], diag_kind="kde")
plt.suptitle("Impact des stars principales sur IMDB_Rating et Gross", y=1.02)
plt.show()
```

<h3 style="text-align:left; color:'brown';">Box Plot of Genres vs. Ratings:</h3>

```python
#Visualize the distribution of IMDB_Rating across different Genres using a Seaborn box plot.
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=df, x="Genre", y="IMDB_Rating")

plt.title("Distribution de IMDB_Rating par Genre")
plt.xlabel("Genre")
plt.ylabel("IMDB Rating")
plt.xticks(rotation=90, fontsize=4)   
plt.tight_layout()
plt.show()
```

<h3 style="text-align:left; color:'brown';">Correlation Heatmap:</h3>

```python
#Create a heatmap using Seaborn to identify correlations between numerical columns like IMDB_Rating, Meta_score, No_of_votes, and Gross.  
```

```python
numeric_cols = ["IMDB_Rating", "Meta_score", "No_of_Votes", "Gross"]

 
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(8,6))

# Heatmap
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Heatmap of Numerical Columns")
plt.show()
```
