<h1 style="text-align:left; color:green;"> Exercise 2: Dogs</h1>

```python
class Dog:
    def __init__(self, name, age, weight):
        self.name=name
        self.age=age
        self.weight=weight

    def bark(self):
        return f'{self.name} is barking'

    def run_speed(self):
        return self.weight / self.age * 10

    def fight(self, other_dog):
        if self.run_speed()*self.weight > other_dog.run_speed()*other_dog.weight:
            return f'{self.name} is the winner'
        elif self.run_speed()*self.weight < other_dog.run_speed()*other_dog.weight:
            return f'{other_dog.name} is the winner'
        else: 
            return f'{other_dog.name} and {self.name} are the twice winners' 
         
    
```

Step 2: Create Dog Instances

```python
Sloughi = Dog("alpha", 12, 10)
Azawakh = Dog("gamma", 8, 8)
Labrador = Dog("Beta", 14, 3)
```

```python
Labrador.fight(Azawakh)
```

<h1 style="text-align:left; color:green;"> Exercise 4: Family and Person Classes</h1>

```python
class Person:
    def __init__(self, first_name, age):
        self.first_name=first_name
        self.age=age
        self.last_name=""
        
    def is_18(self):
        if self.age >= 18:
            return True
        else:
            return False
```

```python
class Family:
    def __init__(self, last_name):
        self.last_name=last_name
        self.members=[]
    
        
    def born(self, first_name, age):
        person = Person(first_name, age)
        person.last_name=self.last_name
        self.members.append(person)

    def check_majority(self, first_name):
        for i in self.members:
            if i.first_name == first_name:
                 
                if i.is_18():
                    print("You are over 18, your parents Zahia and Miloud accept that you will go out with your friends")
                else:
                    print("Sorry, you are not allowed to go out with your friends.")

    def family_presentation(self):
        print(f"The family's last name is {self.last_name}")
        for person in self.members:
            print(f"The person's first name is {person.first_name} and his age is {person.age}")
```

```python
family = Family("KADIRI")
```

```python
family.born("Fatima Ezzahra", 24)
family.born("Ismail", 29)
family.born("Imane", 17)
family.born("Kabira", 27)
 
```

```python
family.check_majority("Fatima Ezzahra")
family.check_majority("Ismail")
family.check_majority("Imane")
family.check_majority("Kabira")
 
```
