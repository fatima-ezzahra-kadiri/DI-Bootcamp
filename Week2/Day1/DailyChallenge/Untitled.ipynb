<h3 style="text-align:left; color:green;">Step 7: Implement the get_short_info Method</h3>

```python
class Farm():
    def __init__(self, farm_name):
        self.name=farm_name
        self.animals={}

    def add_animal(self, animal_type, count=1):
        if animal_type in self.animals:
            self.animals[animal_type]+=count
        else:
            self.animals[animal_type]=count

    def get_info(self):
        print(f"{self.name}'s farm")
        for element, count in self.animals.items():
            print(f"{element} : {count}\n")
        print("     E-I-E-I-0!")

    def get_animal_types(self):
        list_animal = []
        for element in self.animals:
            list_animal.append(element)
        return sorted(list_animal)

    def get_short_info(self):
        list_animals = self.get_animal_types()
        new_list_animals=[]
        for element in list_animals:
            if self.animals[element]>1:
                element+="s"
            new_list_animals.append(element)
            
        # string = ""
        # for i in range(len(new_list_animals)-2):
        #     string += new_list_animals[i]
        #     string += ", "
        # string=string+new_list_animals[len(new_list_animals)-2]+" and "+new_list_animals[len(new_list_animals)-1]+"."
        # print(f"{self.name}'s farm has {string}")

        if len(new_list_animals)==1:
            string = new_list_animals[0]
        else:
            string=", ".join(new_list_animals[:-1])+" and "+new_list_animals[-1]
        return f"{self.name}'s farm has {string}."
```

```python
macdonald = Farm("McDonald")
macdonald.add_animal('cow', 5)
macdonald.add_animal('sheep')
macdonald.add_animal('sheep')
macdonald.add_animal('goat', 12)
macdonald.add_animal('fox')

macdonald.get_short_info()
```

```python
class Farm():
    def __init__(self, farm_name):
        self.name=farm_name
        self.animals={}

    def add_animal(self, **animals_type):
        for animal, count in animals_type.items():
            if animal in self.animals:
                self.animals[animal]+=count
            else:
                self.animals[animal]=count

macdonald = Farm("McDonald")
macdonald.add_animal(cow=5, sheep=2, goat=12)
```
