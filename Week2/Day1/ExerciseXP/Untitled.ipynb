<h1 style="text-align:left; color:green;">Exercise 1: Cats</h1>

```python
class cats():
    def __init__(self,name,age):
        self.name=name
        self.age=age

cat1 = cats("Timoune",6)
cat2 = cats("Honey",12)
cat3 = cats("Simba",2)

def campare_cats(a,b,c):
    if a.age < b.age:
        old = b
        temp = b.age
    else:
        old = a
        temp = a.age
    if temp < c.age:
        old = c
    return old

oldestcat = campare_cats(cat1,cat2,cat3)
print(f"The oldest cat is {oldestcat.name}, and is {oldestcat.age} years old.")

oldestcat = campare_cats(cat1,cat3,cat2)
print(f"The oldest cat is {oldestcat.name}, and is {oldestcat.age} years old.")

oldestcat = campare_cats(cat2,cat1,cat3)
print(f"The oldest cat is {oldestcat.name}, and is {oldestcat.age} years old.")

oldestcat = campare_cats(cat2,cat3,cat1)
print(f"The oldest cat is {oldestcat.name}, and is {oldestcat.age} years old.")

oldestcat = campare_cats(cat3,cat2,cat1)
print(f"The oldest cat is {oldestcat.name}, and is {oldestcat.age} years old.")

oldestcat = campare_cats(cat3,cat1,cat2)
print(f"The oldest cat is {oldestcat.name}, and is {oldestcat.age} years old.")
```

<h1 style="text-align:left; color:green;">Exercise 2 : Dogs</h1>

```python
class Dog:
    def __init__(self, name, height):
        self.name=name
        self.height=height

    def bark(self):
        print(f"{self.name} goes woof!")

    def jump(self):
        print(f"{self.name} jumps {(self.height)*2} cm high!")

davids_dog = Dog("Labrador", 500)
sarahs_dog = Dog("Quik", 700)

print(f"the name of davids_dog is {davids_dog.name} and its height is {davids_dog.height}\nthe name of davids_dog is {sarahs_dog.name} and its height is {sarahs_dog.height}")
davids_dog.bark()
davids_dog.jump()
Dog.bark(davids_dog)
Dog.jump(davids_dog)

def camareDogs(a,b):
    if a.height>=b.height:
        BigDog = a.name
    else:
        BigDog = b.name
    print(f"The Dog with the Biggest size is: {BigDog}")

camareDogs(davids_dog,sarahs_dog)
```

<h3 style="text-align:left; color:purple;">Step 1: Define the Zoo Class</h3>

```python
class Zoo():
    def __init__(self, zoo_name):
        self.zoo_name=zoo_name
        self.animals=[]
      

    def add_animal(self, new_animal):
        if new_animal not in self.animals:
            (self.animals).append(new_animal)

    def get_animals(self):
        for i in self.animals:
            print(f"{i}\n")

    def sell_animal(self, animal_sold):
        if animal_sold in self.animals:
            (self.animals).remove(animal_sold)

 
    def sort_animals(self):
        (self.animals).sort()
        dictionary = {}
        for animal in self.animals:
            fl = animal[0].upper()
            if fl not in dictionary:
                dictionary[fl]=[]
            dictionary[fl].append(animal)
        return dictionary

    def get_groups(self):
        print("the grouped animals")
        for key, value in (self.sort_animals()).items():
            print(f"{key} : {value}\n")

        print("///////////////////////////////////////////////////////////")
        print(f"{self.sort_animals()}")
        print(self.sort_animals())
         
    
    
        
```

```python
Z1 = Zoo("Fantaine")
```
