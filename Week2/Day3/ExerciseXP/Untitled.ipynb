### ðŸŒŸ Exercise 1: Currencies

```python
class Currency:
    def __init__(self, currency, amount):
        self.currency = currency
        self.amount = amount

    def __str__(self):
        return f'{self.amount} {self.currency}s'

    def __repr__(self):
        return self.__str__()

    def __int__(self):
        return  self.amount

    def __add__(self, other):
        if isinstance(other, int):
            return self.amount +  other

        if isinstance(other, Currency):
            if self.currency != other.currency:
                raise TypeError(
                    f"Cannot add between Currency type {self.currency} and {other.currency}"
                )
            return self.amount + other.amount

        raise TypeError("Unsupported type for addition")


    def __iadd__(self, other):
         if isinstance(other, int):
             self.amount += other
             return self

         if isinstance(other, Currency):
             if self.currency != other.currency:
                raise TypeError(
                    f"Cannot add between Currency type {self.currency} and {other.currency}"
                )
             
             self.amount += other.amount
             return self

         raise TypeError("Unsupported type for addition")
    
```

```python
c1 = Currency('dollar', 5)
c2 = Currency('dollar', 10)
c3 = Currency('shekel', 1)
c4 = Currency('shekel', 10)
```

```python
from datetime import datetime

def display_diff_time_until_first_jun():
    current_date_and_time = datetime.now()
    next_year = current_date_and_time.year + 1
    new_date_and_time = datetime(next_year, 1, 1)
    time_difference = new_date_and_time - current_date_and_time
    print(time_difference)

display_diff_time_until_first_jun()
```

### ðŸŒŸ Exercise 6: Birthday and minutes

```python
from datetime import datetime

def seconds_lived_in_life():
    birthdate = str(input("enter your birthdate please (YYYY-MM-DD): "))
    birthdate = datetime.strptime(birthdate, "%Y-%m-%d")
    current_date_and_time = datetime.now()
    difference = current_date_and_time  - birthdate
    minutes = difference.total_seconds() / 60
    print(f'you have lived approximatively {minutes} minutes.') 
    
seconds_lived_in_life()    
```

### ðŸŒŸ Exercise 7: Faker Module

```python
from faker import Faker
```

```python
def funct(nb_users):
    fake=Faker()
    
    for i in range(nb_users):
        user = {
            "name" : fake.name(),
            "address" : fake.address(),
            "language_code" : fake.language_code()
        }
        users.append(user)
        
 
```
