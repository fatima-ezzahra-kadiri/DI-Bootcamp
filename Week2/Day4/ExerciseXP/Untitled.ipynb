<h1 style="text-align:left; color:green;">Exercise 1: Random Sentence Generator</h1>

```python
 

def get_words_from_file(path_file):
    try:
        with open(path_file, "r") as file:
          content = file.read()
          words = content.split()
          return words
    except FileNotFoundError:
        print("Error: the word file was not found.")
        return []
    except Exception as e:
        print(f"An unexpected error occured: {e}")
        return []
        
 
```

```python
import random
def get_random_sentence(length):
    list = get_words_from_file(r"C:\Users\pc\Desktop\Bootcamp\Week2\Day4\ExerciseXP\words.txt")
    sentence_words = [random.choice(list) for i in range(length)]
    sentence = " ".join(sentence_words).lower()
    return sentence
        
```

```python
get_random_sentence(3)
```

```python
def main():
    print ("Purpose : Generate a random sentence of a specified length from a word list.")
     
    try:
        length = int(input("enter the sentence's length taht you want: "))
        if length < 2 or length > 20 :
            print ("error : The sentence length must be between 2 and 20 (inclusive)")
            return
    except ValueError:
        print("error : enter a valid integer between 2 and 20.")
        return

    sentence = get_random_sentence(length)
    print(sentence)
    
if __name__ == "__main__":
    main()
```
