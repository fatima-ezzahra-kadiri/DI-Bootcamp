```python
import numpy as np
import pandas as pd
low = -7
high = 42
n_months = 12
n_cities = 10
size = (n_cities, n_months)


data= np.random.uniform(low, high, size)

index = [f"city_{i}" for i in range(10)]
index
columns = [f"month_{i}" for i in range(1, 13)]
columns
df = pd.DataFrame(data, index, columns)
df
```

<h5 style="text-align:left; color:green;">2. Data Analysis</h5>

```python
annual_average_temperature = df.mean("columns")
annual_average_temperature
```

```python
print("the annual average temperature for each city is :")
annual_average_temperature = df.mean(axis=1)
annual_average_temperature
```

```python
print(f"the city with the highest average temperature is {annual_average_temperature.idxmax()}")
 
```

```python
print(f"the city with the lowest average temperature is {annual_average_temperature.idxmin()}")
 
```

<h5 style="text-align:left; color:green;">2. Data Analysis</h5>

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,6))
for city in df.index:
    plt.plot(df.columns, df.loc[city], marker='o', label=city)

plt.title("Monthly Temperatures per City")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.legend()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.barplot(x=annual_average_temperature.index, y=annual_average_temperature.values, color="yellow")
plt.title("Annual Average Temperature per City")
plt.ylabel("Temperature (C)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Palette de couleurs pour toutes les villes
num_cities = len(df.index)
colors = sns.color_palette("tab10", n_colors=num_cities)

# Crer deux graphes l'un au-dessous de l'autre
fig, axes = plt.subplots(2, 1, figsize=(12, 12), sharex=False)

# Graphique 1 : tempratures mensuelles
for i, city in enumerate(df.index):
    axes[0].plot(df.columns, df.loc[city], marker='o', color=colors[i], label=city)
axes[0].set_title("Monthly Temperatures per City")
axes[0].set_ylabel("Temperature (C)")
axes[0].legend(loc='upper left', bbox_to_anchor=(1,1))

# Graphique 2 : moyenne annuelle par ville
sns.barplot(
    x=annual_average_temperature.index,
    y=annual_average_temperature.values,
    hue=annual_average_temperature.index,  # hue factice
    dodge=False,
    palette=colors,
    legend=False,
    ax=axes[1]
)
axes[1].set_title("Annual Average Temperature per City")
axes[1].set_ylabel("Temperature (C)")
axes[1].set_xlabel("City")

plt.tight_layout()
plt.show()
```



---> City_8 shows the highest average temperature (22.64C), mainly due to several months exceeding 30C, with a peak above 40C in month 5. Despite this, it also experiences a sharp drop close to 3C in month 7, reflecting strong temperature variability.

---> City_0 also has a relatively high yearly average, with stable temperatures between 15C and 35C from months 1 to 10. However, it records the lowest temperature of all cities in month 11, reaching approximately 8C.

---> City_6 is the coldest city, with a low average of 7.98C and temperatures below 15C in most months. Although it shows occasional warm spikes around 3035C in months 4 and 11, its overall climate remains predominantly cold.
