# 🌟 Exercise 2: Descriptive Statistics

```python
from scipy import stats
import numpy as np

data = [12, 15, 13, 12, 18, 20, 22, 21]

mean = np.mean(data)
median = np.median(data)
variance = np.var(data, ddof=1)
std_dev = np.std(data, ddof=1)

print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

# 🌟 Exercise 3: Understanding Distributions)”.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

mean = 50
std_dev = 10
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = norm.pdf(x, mean, std_dev)

plt.plot(x, y)
plt.title("Normal Distribution (mean=50, std=10)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
```

# 🌟 Exercise 4: T-Test Application

```python
import numpy as np
from scipy import stats

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_value = stats.ttest_ind(data1, data2)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

# 🌟 Exercise 5: Linear Regression Analysis

```python
house_sizes = [50, 70, 80, 100, 120]  # in square meters
house_prices = [150000, 200000, 210000, 250000, 280000] 
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

print("Slope:", slope)
print("Intercept:", intercept)

# Predict price for a 90 sqm house
predicted_price = slope * 90 + intercept
print("Predicted price for 90 sqm:", predicted_price)
```

Questions:

Slope interpretation: For each additional square meter, the house price increases by the slope amount.

Intercept interpretation: Estimated base price when house size is zero.

# 🌟 Exercise 6: Understanding ANOVA

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
```

```python
from scipy import stats

f_stat, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_stat)
print("P-value:", p_value)
```

- P-value < 0.05, the fertilizers have significantly different effects on plant growth.

# 🌟 Exercise 8: Correlation Coefficients (Optional)

```python
import pandas as pd
from scipy import stats

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

pearson_corr, _ = stats.pearsonr(data['age'], data['income'])
spearman_corr, _ = stats.spearmanr(data['age'], data['income'])

print("Pearson correlation:", pearson_corr)
print("Spearman correlation:", spearman_corr)
```
