# 🌟 Exercise 1 : Matrix Operations

```python
import numpy as np

# Create a 3x3 matrix
matrix = np.array([
    [2, 4, 1],
    [3, 5, 7],
    [1, 0, 6]
])

print("Matrix:\n", matrix)

# Determinant
determinant = np.linalg.det(matrix)
print("\nDeterminant:", determinant)

# Inverse
inverse_matrix = np.linalg.inv(matrix)
print("\nInverse Matrix:\n", inverse_matrix)
```

# 🌟 Exercise 3 : Date Manipulation

```python
import numpy as np

# Create dates for January 2023
dates = np.arange('2023-01-01', '2023-02-01', dtype='datetime64[D]')

print("Original Dates:\n", dates)

# Convert format YYYY/MM/DD
formatted_dates = [str(date).replace('-', '/') for date in dates]

print("\nFormatted Dates:\n", formatted_dates)
```

# 🌟 Exercise 6 : Basic Hypothesis Testing

Hypothesis

- H₀ (Null Hypothesis): The training program has no effect on productivity.

- H₁ (Alternative Hypothesis): The training program improves productivity.

```python
import numpy as np

# Data
productivity_before = np.random.normal(loc=50, scale=10, size=30)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Mean productivity
mean_before = np.mean(productivity_before)
mean_after = np.mean(productivity_after)

print("Mean before training:", mean_before)
print("Mean after training:", mean_after)

# Difference
difference = mean_after - mean_before

print("Average productivity improvement:", difference)

# Simple decision rule
if difference > 0:
    print("Conclusion: Training program improved productivity.")
else:
    print("Conclusion: No improvement detected.")
```
