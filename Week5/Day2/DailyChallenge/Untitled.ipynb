# Data Import and Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("global_power_plant_database.csv")

df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
# Replace missing capacity with median
df['capacity_mw'] = df['capacity_mw'].fillna(df['capacity_mw'].median())

# Drop rows where country is missing
df = df.dropna(subset=['country'])
```

```python
df['capacity_mw'] = pd.to_numeric(df['capacity_mw'], errors='coerce')

df['commissioning_year'] = pd.to_numeric(df['commissioning_year'], errors='coerce')
```

# Statistical Analysis

```python
fuel_stats = df.groupby("primary_fuel")["capacity_mw"].agg([
    "mean",
    "median",
    "std",
    "count"
])

fuel_stats
```

Hypothesis Testing (Coal vs Gas)
Hypothesis

H₀ : Mean power output is the same for Coal and Gas plants

H₁ : Mean power output is different

```python
from scipy import stats

coal_capacity = df[df["primary_fuel"] == "Coal"]["capacity_mw"]
gas_capacity = df[df["primary_fuel"] == "Gas"]["capacity_mw"]

t_stat, p_value = stats.ttest_ind(coal_capacity, gas_capacity, nan_policy='omit')

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject H0 --> Significant difference in power output.")
else:
    print("Fail to reject H0 --> No significant difference.")
```

# Matrix Operations in Real-World Context

```python
features = df[["capacity_mw","latitude","longitude"]].dropna()

matrix = features.to_numpy()
```

```python
cov_matrix = np.cov(matrix.T)

print(cov_matrix)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:", eigenvectors)
```

Interpretation

- Eigenvalues show the variance captured by each principal direction.
- Eigenvectors indicate relationships between variables (capacity, location).
- This concept is used in PCA (Principal Component Analysis) for dimensionality reduction.

```python
x = df["capacity_mw"].dropna()
y = df["latitude"].dropna()

coeff = np.polyfit(x[:len(y)], y, 1)

line = np.poly1d(coeff)

plt.scatter(x[:len(y)], y)
plt.plot(x[:len(y)], line(x[:len(y)]))

plt.title("Relationship between Capacity and Latitude")

plt.show()
```
